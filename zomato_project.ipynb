### Dropping the Duplicates

```python
df.drop_duplicates(inplace=True)
df.shape
```

## Cleaning the Rate Column

```python
df['rate'].unique()
```

## Function that removes the 'NEW' , '-' and '/5' from the Rate column

```python
def handle_rate_culomn(value):
  if(value=='NEW'or value=='-'):
    return np.nan
  else:
    value= str(value).split('/')    # split the string at "/"
    value= value[0]                 # split function return the list object
    return float(value)             # return the float value


df['rate']= df['rate'].apply(handle_rate_culomn)
df['rate'].head()
```

## Filling Null Values in the rate Column with Mean values of Rate column

```python
df.rate.isnull().sum()    #finding how much null values is there
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

## Dropping Null Values

```python
df.dropna(inplace = True)
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2plates','listed_in(type)':'type','listed_in(city)':'city'},inplace=True)  # Changing the name of few column
df.head()
```

```python
df['city'].unique()
```

```python
df['location'].unique()
```

 As location column contain more unique values then city column we can drop the city column because both the column gives similar location

---



```python
df= df.drop(['city'],axis=1)
```

```python
df['cost2plates'].unique()
```

## Removing the , from cost2plates column

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)

df['cost2plates'] = df['cost2plates'].apply(handlecomma)
df['cost2plates'].unique()
```

```python
df.head()
```

## Cleaning Rest Type Column

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

## Making Rest Types less than 1000 in frequency as others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

## Cleaning Location Column

```python
location = df['location'].value_counts(ascending  = False)

location_lessthan300 = location[location<300]



def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

## Cleaning Cuisines Column

```python
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
